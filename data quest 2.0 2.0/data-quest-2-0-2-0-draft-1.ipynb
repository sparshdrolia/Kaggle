```python
#make one hot vectors
X_new = pd.get_dummies(X, columns = ["Profession","Status","edu","Irregular","residence","prev_diagnosed","communication","Month","side_effects"],
                             prefix=[ "Prof","Status","edu","Irr","resi","prevD","comm","Month","SideE"])

testF_new = pd.get_dummies(testF, columns = ["Profession","Status","edu","Irregular","residence","prev_diagnosed","communication","Month","side_effects"],
                             prefix=[ "Prof","Status","edu","Irr","resi","prevD","comm","Month","SideE"])


X_new.info()
```

```python
from sklearn.utils import resample
df_majority = X_new[X_new.Y==0]
df_minority = X_new[X_new.Y==1]

df_minority_upsampled = resample(df_minority, 
                                 replace=True,     # sample with replacement
                                 n_samples=14000,    # to match majority class
                                 random_state=123)

df_upsampled = pd.concat([df_majority, df_minority_upsampled])
#df_upsampled.Y.value_counts()

df_majority_downsampled = resample(df_majority, 
                                 replace=False,    # sample without replacement
                                 n_samples=15000,     # to match minority class
                                 random_state=123)

df_downsampled = pd.concat([df_majority_downsampled, df_minority_upsampled])
df_downsampled.Y.value_counts()
```
