```python
for i in range(0, len(X)):
    if X['Fare'][i] <= 7.91:
        X['Fare'][i] = 0
    elif (X['Fare'][i] > 7.91) & (X['Fare'][i] <=14.454):
        X['Fare'][i] = 1
    elif (X['Fare'][i] > 14.454) & (X['Fare'][i] <=31):
        X['Fare'][i] = 2
    elif X['Fare'][i] > 31:
        X['Fare'][i] = 3
        
    if X['Age'][i] <= 16:
        X['Age'][i] = 0
    elif (X['Age'][i] > 16) & (X['Age'][i] <= 32):
        X['Age'][i] = 1
    elif (X['Age'][i] > 32) & (X['Age'][i] <= 48):
        X['Age'][i] = 2
    elif (X['Age'][i] > 48) & (X['Age'][i] <= 64):
        X['Age'][i] = 3
    elif X['Age'][i] > 64:
        X['Age'][i] = 4
    
    if X['Sex'][i] == "male":
        X['Sex'][i] = 0
    elif X['Sex'][i] == "female":
        X['Sex'][i] = 1
        
    if X['Embarked'][i] == "C":
        X['Embarked'][i] = 0
    elif X['Embarked'][i] == "S":
        X['Embarked'][i] = 1
    elif X['Embarked'][i] == "Q":
        X['Embarked'][i] = 2
```

```python
for i in range(0, len(test_data)):
    if test_data['Fare'][i] <= 7.91:
        test_data['Fare'][i] = 0
    elif (test_data['Fare'][i] > 7.91) & (test_data['Fare'][i] <=14.454):
        test_data['Fare'][i] = 1
    elif (test_data['Fare'][i] > 14.454) & (test_data['Fare'][i] <=31):
        test_data['Fare'][i] = 2
    elif test_data['Fare'][i] > 31:
        test_data['Fare'][i] = 3
        
    if test_data['Age'][i] <= 16:
        test_data['Age'][i] = 0
    elif (test_data['Age'][i] > 16) & (test_data['Age'][i] <= 32):
        test_data['Age'][i] = 1
    elif (test_data['Age'][i] > 32) & (test_data['Age'][i] <= 48):
        test_data['Age'][i] = 2
    elif (test_data['Age'][i] > 48) & (test_data['Age'][i] <= 64):
        test_data['Age'][i] = 3
    elif test_data['Age'][i] > 64:
        test_data['Age'][i] = 4
    
    if test_data['Sex'][i] == "male":
        test_data['Sex'][i] = 0
    elif test_data['Sex'][i] == "female":
        test_data['Sex'][i] = 1
        
    if test_data['Embarked'][i] == "C":
        test_data['Embarked'][i] = 0
    elif test_data['Embarked'][i] == "S":
        test_data['Embarked'][i] = 1
    elif test_data['Embarked'][i] == "Q":
        test_data['Embarked'][i] = 2
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
import keras.utils.np_utils.to_categorical
```

```python
print(np.shape(X))
```

```python
model = Sequential()

model.add(Dense(500, input_dim = 891))
model.add(Activation('relu'))

model.add(Dense(200))
model.add(Activation('relu'))

model.add(Dense(70))
model.add(Activation('relu'))

model.add(Dense(30))
model.add(Activation('relu'))

model.add(Dense(15))
model.add(Activation('sigmoid'))

model.add(Dense(5))
model.add(Activation('tanh'))

model.add(Dense(1))
model.add(Activation('sigmoid'))
```

```python
model.compile(optimizer = 'rmsprop', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(X, y, epochs=10, batch_size=50)
```
